```python
# 4. Devolver los restaurantes que lograron una puntuación superior a 80 pero inferior a 100

# filtro = {"grades.score": {"$gt": 80}, "grades.score": {"$lt": 100}}

filtro = {'grades': {'$elemMatch': {'score': {'$gt': 80, '$lt': 100}}}}
proyeccion = {"name":1, "grades":1, "_id": 0}
list(restaurantes.find(filtro, proyeccion))
```

```python
# 5. Devolver los restaurantes que se ubican en un valor de latitud inferior a -95.754168
filtro ={'address.coord.0': {'$lt': -95.754168}}
proyeccion = {'_id':0, 'name':1}

list(restaurantes.find(filtro, proyeccion))
```

```python
# 6. Devolver los restaurantes que no preparan cocina americana y lograron una puntuación superior a 70 y se ubicaron en una longitud inferior a -65.754168. 
# Nota: Realice esta consulta sin usar el operador $and
filtro = {'cuisine': {'$nin': ['American ']}, # '$ne': 'American '
          'grades.score' : {'$gt': 70},
          'address.coord.0': {'$lt': -65.754168}
          }
# filtro = {"cuisine": {"$not": {"$eq":"American "}}, 
#           "grades.score": {"$gt": 70}, 
#           "address.coord.0": {"$lt": -65.754168}
#         }
proyeccion = {'_id':0, 'cuisine':1, 'grades':1, 'address':1, 'name':1}
    
list(restaurantes.find(filtro, proyeccion))
```

```python
# 7. Devolver los restaurantes que no preparan cocina americana y lograron un punto de calificación 'A' que no pertenece al distrito de Brooklyn. 
# El documento debe mostrarse según la cocina en orden descendente.
filtro = {'cuisine': {'$nin': ['American ']},
          'grades.grade': 'A',
          'borough': {'$nin': ['Brooklin']}}
proyeccion = {'_id':0, 'borough':1, 'name':1, 'cuisine':1}
list(restaurantes.find(filtro, proyeccion).sort('cuisine', -1))
```

```python
# 8. Devolver los restaurantes que pertenecen al distrito Bronx y preparan platos americanos o chinos
filtro = {'borough': 'Bronx',
          '$or': [{'cuisine': 'American '},
                  {'cuisine': 'Chinese'}]

}
proyeccion = {'_id':0}
    
list(restaurantes.find(filtro, proyeccion))
```

```python
# 9. Devolver ID del restaurante, nombre, distrito y la cocina para aquellos restaurantes que pertenecen al distrito de Staten Island o Queens o Bronx o Brooklyn
filtro = {'$or': [{'borough': 'Saten Island'}, 
                  {'borough': 'Queens'},
                  {'borough': 'Bronx'},
                  {'borough': 'Brooklin'}]}
proyeccion = {'_id':0, 'restaurant_id': 1, 'name':1, 'borough':1, 'cuisine':1}
    
list(restaurantes.find(filtro, proyeccion))
```

```python
# 10. Devolver ID del restaurante, nombre, distrito y la cocina de aquellos restaurantes que lograron una puntuación que no supere los 10

filtro = {'grades.score': {'$lt': 10}}
proyeccion = {'_id':0, 'restaurant_id': 1, 'name':1, 'borough':1, 'cuisine':1}
    
list(restaurantes.find(filtro, proyeccion))
```

```python
from datetime import datetime, timezone

# Fecha en formato ISO con zona horaria UTC
fecha_iso = "2014-08-11T00:00:00Z"

# Convertir la fecha ISO a un objeto datetime, asegurando que esté en UTC
fecha_objetivo = datetime.fromisoformat(fecha_iso.replace("Z", "+00:00"))

# Convertir el objeto datetime a timestamp en milisegundos
timestamp_ms = int(fecha_objetivo.timestamp() * 1000)

filtro = {'grades.grade': 'A',
          'grades.score': 11,
          'grades.date.$date': timestamp_ms}
         
proyeccion = {'_id':0, 'restaurant_id': 1, 'name':1, 'grades':1}
    
list(restaurantes.find(filtro, proyeccion)) 
```

```python
# 11. Devolver ID del restaurante, el nombre y las calificaciones del restaurante para aquellos restaurantes que obtuvieron una calificación de "A" 
# y obtuvieron un puntaje de 11 en una fecha ISO "2014-08-11T00: 00: 00Z" entre muchas fechas de encuesta
filtro = {'grades.grade': 'A',
          'grades.score': 11,
          'grades.date.$date': timestamp_ms}
         
proyeccion = {'_id':0, 'restaurant_id': 1, 'name':1, 'grades':1}
    
list(restaurantes.find(filtro, proyeccion))
```

```python
# 12. Devolver ID del restaurante, nombre, dirección y ubicación geográfica del restaurante de 
# aquellos donde el segundo elemento de la matriz coord contiene un valor que es más de 42 y hasta 52
filtro = {'address.coord.1': {'$gt': 42, '$lt': 52}}
proyeccion = {'_id':0, 'restaurant_id': 1, 'name':1, 'address.building':1, 'address.coord':1}
    
list(restaurantes.find(filtro, proyeccion))
```
