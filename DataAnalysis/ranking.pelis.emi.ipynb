```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
```

```python
rutilla = 'https://www.filmaffinity.com/es/ranking.php?rn=ranking_fa_movies'
driver = webdriver.Chrome()
url = rutilla
driver.get(url)

boton_consentimiento = '//*[@id="qc-cmp2-ui"]/div[2]/div/button[2]'
boton_consentimiento = driver.find_element(By.XPATH, boton_consentimiento)
boton_consentimiento.click()

import time

for clicks in range(8):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    path_boton_pabajo = '//*[@id="mt-content-cell"]/div[2]/div[3]/i[1]'
    boton_pabajo = driver.find_element(By.XPATH, path_boton_pabajo)
    boton_pabajo.click()
    time.sleep(2)

soup = bs(driver.page_source, 'lxml')
```

```python
titulo=soup.select('div.mc-title')
lista_titulos= []
for cosa in titulo:
    lista_titulos.append(cosa.text)

year= soup.select('span.mc-year')
lista_year= []
for cosa in year:
    lista_year.append(cosa.text)

posicion = soup.select('li.position')
lista_posicion= []
for cosa in posicion:
    lista_posicion.append(cosa.text)

rating = soup.select('div.avg-rating')
lista_rating= []
for cosa in rating:
    lista_rating.append(cosa.text.replace(",","."))

director = soup.select('div.mc-director')
lista_director= []
for cosa in director:
    lista_director.append(cosa.text)

cast = soup.select('div.mc-cast')
lista_cast = []

for cosa in cast:
    actores = cosa.text.split(",")[:3]  
    actores = [actor.strip() for actor in actores]
    actores_string = ", ".join(actores)
    lista_cast.append(actores_string)
lista_cast
```

```python
datos = {'titulo': lista_titulos, 'director': lista_director, 'cast': lista_cast, 'year': lista_year, 'posicion': lista_posicion, 'rating': lista_rating}
df = pd.DataFrame(datos)
df = df.iloc[:-20]
```
