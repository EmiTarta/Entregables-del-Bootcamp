8. Crea un gr치fico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sof칤a mide 190 cm, Carmen mide 170 cm. Las barras ser치n verdes.

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
nombres = ['Ana', 'Luis', 'Pedro', 'Sof칤a', 'Carmen']
alturas = [160, 180, 175, 190, 170]

barplot = plt.bar(nombres, alturas, color = 'green')
plt.xlabel('Personas')
plt.ylabel('Altura (cm)')

plt.title('Alturas grupo de amigos') 

for bar in barplot:
    altura = bar.get_height()  
    x = bar.get_x() + bar.get_width() / 2 
    plt.text(x, altura, f'{altura} cm', ha='center', va='bottom') 

plt.show()
```

17. Otro departamento B ha vendido en esos a침os 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gr치fica las dos l칤neas. B es una l칤nea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
data = {
    'Year' : [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A' : [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B' : [2000, 3100, 5000, 4000, 6000]}
df = pd.DataFrame(data)

plt.style.use('default')

fig, ax = plt.subplots()
line_a = ax.plot(df['Year'], df['Sold_items_A'], linestyle='--', color='green', label='Ventas A')
plt.xticks(df['Year'].astype(int))
line_b, = ax.plot(df['Year'], df['Sold_items_B'], linestyle=':', color='red', label='Ventas B')
ax.legend(loc='lower right')

plt.show()
```
