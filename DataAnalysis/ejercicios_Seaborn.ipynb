```python
# 19. Veamos la función melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
# Dado un DataFrame, melt recibe varios parámetros:
# * qué columnas del DataFrame se quedan igual
# * qué nombres de columnas del dataframe pasan a ser una columna nueva
# * cómo queremos llamar esas nuevas columnas

# Primero recuerda el Dataframe df_grass_fire_water con las primeras 5 filas
```

```python
grass_fire_water.head(5)
```

```python
grass_fire_water.columns
```

```python
# 20. Queremos dejar igual las variables 'Name' y 'Type 1'
# Queremos añadir una columa nueva cuyos valores sean las columnas que queremos apilar y que se llame 'New Column'
# Escribimos df_grass_fire_water.melt(id_vars=['Name', 'Type 1'], value_vars=[...], var_name='New Column')
```

```python
gfw_melt=grass_fire_water.melt(id_vars=['Name', 'Type 1'], value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk','Sp. Def', 'Speed'], var_name='New Column')
```

```python
gfw_melt.head(5)
```

```python
# 21. Pintamos el swarmplot con el melted y entendemos la información
```

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(x='New Column', y='value', hue='Type 1', data=gfw_melt, palette=pokemon_colores, dodge=False)

plt.title('Distribución de Valores por Tipo de Pokémon')
plt.ylabel('Valores')
plt.xlabel('Características')
plt.legend(title='Tipo de Pokémon')
plt.tight_layout()
plt.show()
```

```python
# 22. Basándote en el gráfico anterior, define el tamaño de la figura en (10,6) (usa plt)
# En el swarmplot añade el argumento dodge a True y usa la anterior paleta de colores Pokemon

# Dodge en inglés es "esquivar", ¿qué diferencia ves?

# Usando plt 
# fija el límite del eje Y entre 0 y 200
# fija la leyenda con bbox_to_anchor en upper left. Pista: upper left es el código 2 y se usa el parámetro loc.
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='New Column', y='value', hue='Type 1', data=gfw_melt, palette=pokemon_colores, dodge=True)
plt.ylim(0, 200)

plt.title('Distribución de Valores por Tipo de Pokémon')
plt.ylabel('Valores')
plt.xlabel('Melted Variables')
plt.legend(title='Tipo de Pokémon', bbox_to_anchor=(0, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# 26. Usa catplot con el DataFrame original. Muestra el valor de Defense según el 'Type 1', muestra un subplot para legendarios y otro para no legendarios
# y divide los colores según el Type 2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asumiendo que tienes una columna que indica si es legendario o no, llamada 'Legendary'
# Separa el DataFrame en legendarios y no legendarios
legendary_df = pokemon_df[pokemon_df['Legendary'] == True]
non_legendary_df = pokemon_df[pokemon_df['Legendary'] == False]

# Establece el tamaño de la figura
fig = plt.figure(figsize=(10,10))
sns.set(rc={'figure.figsize':(30, 20)})  # Ajusta el tamaño de la figura en Seaborn

# Crea el catplot
g = sns.catplot(
    x='Type 1', 
    y='Defense', 
    hue='Type 2', 
    data=pokemon_df, 
    height=6, 
    aspect=2, 
    col='Legendary',  # Esta columna debe ser booleana
    col_order=[False, True],  # Asegura que 'No Legendario' sea primero
    palette='muted'  # Cambia la paleta según tus preferencias
)

# Ajusta los títulos de los subplots
g.set_titles(col_template="Legendary = {col_name}")

# Cambia el nombre de las columnas para que muestre True o False
for ax, col in zip(g.axes.flat, g.col_names):
    ax.set_title(f"Legendary = {col}")  # Cambia el título del subplot


# Muestra el gráfico
plt.show()

```

```python
# 27. Muestra una función distribución de probabilidad con kdeplot (curvas de nivel) de la defensa respecto al ataque.
# Di un bin 2D donde caigan muchos puntos (ataque, defensa).
```

```python
sns.kdeplot(
    x='Attack', 
    y='Defense', 
    data=pokemon_df, 
    color='blue',      # Color uniforme para todas las líneas
    fill=False,        # No rellenar el área bajo la curva
    thresh=0,          # Incluye todas las densidades
             # Número de niveles de densidad
)

# Etiquetas y título
plt.title('Distribución de Densidad de Ataque vs Defensa')
plt.xlabel('Attack')
plt.ylabel('Defense')

# Muestra el gráfico
plt.show()
```
